```python
import pandas as pd
from string import ascii_uppercase as alphabet
import pickle
```

```python
all_tables = pd.read_html('https://web.archive.org/web/20221115040351/https://en.wikipedia.org/wiki/2022_FIFA_World_Cup')
```

```python
#extract and set all the group table
#clean the dataframe and create a dictionary

dict_table = {}

for letter, i in zip(alphabet, range(12,68,7)):
    df = all_tables[i]
    df.rename(columns={df.columns[1]:'Team'}, inplace = True)
    #df.pop('Qualification') --> delete the Qualification column
    dict_table[f'Group {letter}'] = df
    
```

```python
#export the dictionary in a file
with open('dict_table', 'wb') as output:
    pickle.dump(dict_table, output)
```
